# Expense Tracker

```python
''' User salary add karta hai: +50,000 (Income)
User grocery add karta hai: –7,000 (Expense, Category: Food)
User transport add karta hai: –3,000 (Expense, Category: Transport)
System show karega:
Total Income: 50,000
Total Expenses: 10,000
Remaining Balance: 40,000
Top Expense Category: Food '''

from datetime import datetime as dt

class expense_tracker:
    
    def __init__(self , total_income):
          self.expenses = {
        "food": 0,
        "transport": 0,
        "shopping": 0
     }
        
          self.total_income = total_income 
          print(f'Your Total income is : {self.total_income}')

    
    def add_expense(self ,category , amount ) :
        current_time = dt.now().strftime('%y-%m-%d  %H:%M:%S')
        
        if category in self.expenses:
            self.expenses[category] +=amount
            print(f"Added {amount} to {category} at {current_time}")

        else:
            print(f"{category} is not a valid category!")


    
    def show(self):
        
        for key , val in self.expenses.items():
              print(f'{key} : {val}')

    
        
    def calculate(self):
        total_expense = 0
        Remaining_Balance = 0
        
        for val in self.expenses.values():
            total_expense +=val
             
        Remaining_Balance  = self.total_income - total_expense
        
        if Remaining_Balance > 0:
            return f' Total Expense is {total_expense} \n Remaining Balance is {Remaining_Balance} '
        else :
            return f"Total Expense is {total_expense}\n You have exceeded your budget by {-Remaining_Balance}!"


    
    def top_expense(self):
        
        if not self.expenses:
            return "No expenses recorded"

        category = max(self.expenses , key=self.expenses.get)
        amount = self.expenses[category]
        return f"Highest Expense: {category} = {amount}"
        
              
        
        
            
    

user1 = expense_tracker(50000) 
user1.add_expense('food' ,4000)
user1.add_expense('transport' , 2000)
user1.add_expense('shopping'  , 10000 )
user1.show()
print(user1.calculate())
print(user1.top_expense())

print()
print()


user2 = expense_tracker(100000) 
user2.add_expense('food' ,20000)
user2.add_expense('transport' , 30000)
user2.add_expense('shopping'  , 80000 )
user2.show()
print(user2.calculate())
print(user2.top_expense())

    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "Category": list(user1.expenses.keys()) * 2,
    "Expense": list(user1.expenses.values()) + list(user2.expenses.values()),
    "User": ["User 1"] * 3 + ["User 2"] * 3
})

plt.figure(figsize=(8,4))
sns.barplot(
    x="Category", 
    y="Expense", 
    hue="User", 
    data=df,
    palette="Set2",   
    edgecolor="black" 
)

plt.title("User1 vs User2 Expense Comparison", fontsize=14, fontweight="bold")
plt.xlabel("Expense Category", fontsize=12)
plt.ylabel("Amount ", fontsize=12)
plt.legend(title="Users")
plt.show()
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))


colors1 = sns.color_palette("pastel", len(user1.expenses))  
colors2 = sns.color_palette("Set2", len(user2.expenses))  

# --- Pie chart User 1 ---
ax1.pie(user1.expenses.values(),
        labels=user1.expenses.keys(),
        autopct="%1.1f%%",
        colors=colors1,
        startangle=90,
        wedgeprops={"edgecolor":"white"})
ax1.set_title("Expense Distribution For User 1")

# --- Pie Chart User 2 ---
ax2.pie(user2.expenses.values(),
        labels=user2.expenses.keys(),
        autopct="%1.1f%%",
        colors=colors2,
        startangle=90,
        wedgeprops={"edgecolor":"white"})
ax2.set_title("Expense Distribution For User 2")

plt.tight_layout()
plt.show()

```
